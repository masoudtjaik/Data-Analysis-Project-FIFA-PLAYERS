```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fifa_players.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
columns = [
    'age',
    'positions',
    'overall_rating',
    'potential',
    'value_euro',

    'finishing',
    'dribbling',
    'ball_control',
    'short_passing',
    'long_passing',

    'vision',
    'crossing',
    'curve',

    'acceleration',
    'sprint_speed',
    'agility',
    'balance',
    'stamina',
    'strength',

    'reactions'
]
drop_columns = [col for col in df.columns if col not in columns]
df.drop(drop_columns, axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['value_euro']=df['value_euro'].fillna(df['value_euro'].median())
```

```python
df.isnull().sum()
```

```python
df
```

```python
plt.bar(np.arange(len(df['overall_rating'])),df['overall_rating'], label='Overall Rating')
plt.xlabel('Players')
plt.ylabel('Rating')
plt.legend()
plt.show()
```

```python
sns.histplot(df['overall_rating'])
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
for col in numeric_df.columns:
    plt.figure(figsize=(8,1))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
Q1=np.percentile(df['overall_rating'],25)
Q2=np.percentile(df['overall_rating'],50)
Q3=np.percentile(df['overall_rating'],75)
IQR=Q3-Q1
lower=Q1-1.5 * IQR
upper=Q3+1.5 * IQR
df_clean = df[
    (df['overall_rating'] >= lower) &
    (df['overall_rating'] <= upper)
]
df_clean
```

```python
plt.figure(figsize=(8,1))
sns.boxplot(x=df_clean['overall_rating'])
plt.title('Overall Rating')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(14,6))
sns.heatmap(corr, cmap='coolwarm',annot=True)
plt.title('Correlation Matrix')
plt.show()
```
